# Basic visualizations

## Size class distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='SIZE_CLASS', order=sorted(df['SIZE_CLASS'].dropna().unique()))
plt.title('Distribution of Fire Size Classes')
plt.xlabel('Size Class')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## General Cause of Fire

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='GENERAL_CAUSE', order=df['GENERAL_CAUSE'].value_counts().index)
plt.title('General Cause of Fire')
plt.xlabel('Count')
plt.ylabel('General Cause')
plt.show()
```

## Fires by Year

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='YEAR')
plt.title('Number of Fires per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Fire Start Size vs Detected Size

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='FIRE_FIGHTING_START_SIZE', y='DISCOVERED_SIZE')
plt.title('Fire Fighting Start Size vs Discovered Size')
plt.xlabel('Fire Fighting Start Size (hectares)')
plt.ylabel('Discovered Size (hectares)')
plt.show()
```

## Weather (Temperature vs Wind Speed)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='TEMPERATURE', y='WIND_SPEED', hue='SIZE_CLASS')
plt.title('Temperature vs Wind Speed Colored by Fire Size Class')
plt.xlabel('Temperature (C)')
plt.ylabel('Wind Speed (km/h)')
plt.legend(title='Size Class')
plt.show()
```
